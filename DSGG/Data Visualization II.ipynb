```python
warnings.filterwarnings('ignore')
```

```python
%matplotlib inline
```

```python
data = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.countplot(data['parch'])
```

```python
data['sex'].value_counts().plot(kind="pie", autopct="%.2f")
```

```python
sns.histplot(data['age'], bins=10,kde=True)
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips['tip'])
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips['tip'], hue=tips['sex'])
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips['tip'], hue=tips['sex'], style=tips['smoker'])
```

```python
sns.barplot(x=data['pclass'], y=data['age'])
```

```python
sns.barplot(x=data['pclass'], y=data['age'], hue=data['sex'])
```

```python
sns.heatmap(pd.crosstab(data['pclass'], data['survived']),annot= True )
```
