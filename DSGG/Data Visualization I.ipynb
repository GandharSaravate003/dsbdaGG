```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('titanic')
dataset.head()
```

# Distribution Plots


```python
# Dist-plot
```

```python
import seaborn as sns
sns.displot(x = dataset['age'], bins = 10, kde=True)
```

```python
sns.displot(dataset['age'], bins = 10,kde=False)
```

```python
#b. Joint Plot
```

```python
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind ='scatter')
```

```python
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'hex')
```

```python
#c. The Rug Plot
```

```python
sns.rugplot(dataset['fare'])
```

# Categorical plot

```python
sns.barplot(x='sex', y='age', data=dataset,hue='survived', palette=["blue", "yellow"])
```

```python
sns.barplot(hue='sex', y='age', data=dataset, estimator=np.std,palette=["blue", "yellow"])
```

```python
#b. The Count Plot
```

```python
sns.countplot(x='sex', data=dataset,hue='survived', palette=["blue", "yellow"])
```

```python
#c. The Box Plot
```

```python
sns.boxplot(x='sex', y='age', data=dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived",palette=["blue", "yellow"])
```

```python
#d. The Violin Plot
```

```python
sns.violinplot(x='sex', y='age', data=dataset)
```

```python
sns.violinplot(x='sex', y='age', data=dataset, hue='survived',palette=["blue", "yellow"])
```

# Advanced Plots:

```python
#a. The Strip Plot
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=False,hue='survived',palette=["blue", "yellow"])
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True)
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True, hue='survived',palette=["blue", "yellow"])
```

```python
#b. The Swarm Plot
```

```python
sns.swarmplot(x='sex', y='age', data=dataset)
```

```python
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived',palette=["blue", "yellow"])
```
